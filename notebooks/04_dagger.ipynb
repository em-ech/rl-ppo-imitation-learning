Plain BC suffers covariate shift. DAgger rolls out the current student and queries the expert on the visited states, training on the student's own distribution. This notebook **runs DAgger live** (reduced iterations) via `src.dagger.run_dagger` (the shared core also used by `dagger_run.py`), then shows the precomputed full BC-vs-DAgger comparison.

```python
import sys, os, json
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT)); os.chdir(ROOT)
%matplotlib inline
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from IPython.display import Image, display
from src import seeding, config
seeding.set_seed(0)            # numpy + torch + python random
DEVICE = config.device()
OUT, MODELS, DATA = ROOT/'outputs', ROOT/'models', ROOT/'data'/'demonstrations'
def show(name):
    p = OUT/name
    display(Image(str(p))) if p.exists() else print('precomputed figure not found:', name,
        '- run the corresponding script (see README)')
def results(name):
    p = OUT/name
    return json.load(open(p)) if p.exists() else {}
print('setup ok | seeds set | device', DEVICE)
```

## DAgger live on Walker2d (6 iterations from the submitted expert)

```python
import torch; torch.set_num_threads(1)
from src import dagger
if (MODELS/'ppo_expert_Walker2d-v4'/'best_model.zip').exists():
    out = dagger.run_dagger('Walker2d-v4', n_iters=6, steps_per_iter=5000,
                            bc_epochs=25, seed=0)
    r = out['returns_by_iter']
    plt.plot(range(1,len(r)+1), r, marker='o', color='darkorange')
    plt.xlabel('DAgger iteration'); plt.ylabel('eval return')
    plt.title('DAgger live (Walker2d, 6 iters)'); plt.grid(alpha=.3); plt.show()
    print('DAgger live: start %.0f -> end %.0f (best %.0f)' % (r[0], r[-1], max(r)))
else:
    print('expert missing - run train_expert.py')
```

## Full BC vs DAgger comparison (precomputed, 12 iters x 25 BC epochs)

```python
keys = {'Walker2d-v4':'Walker2d (6043)','Walker2d-v4_generic_backup':'Walker2d (4627)','Ant-v4':'Ant (6293)'}
rows=[]
for k,label in keys.items():
    bc=results(f'bc_results_{k}.json'); dg=results(f'dagger_results_{k}.json')
    if not bc or not dg: continue
    em=bc.get('expert_mean'); lib=bc['library_bc']['mean']; dgr=dg['returns_by_iter']
    rows.append(dict(config=label, expert=round(em), BC_library=round(lib),
        DAgger_final=round(dgr[-1]), DAgger_best=round(max(dgr))))
display(pd.DataFrame(rows))
for k in keys: show(f'bc_vs_dagger_{k}.png')
```

## RQ5 - Does DAgger reduce covariate shift vs plain BC?

**Yes, clearly.** With a fair training budget DAgger matches or beats BC on every config and reaches expert level: Walker2d-6043 6208 (> BC 5719, ~expert), Walker2d-4627 4732, Ant 6564 (> BC 6237). The mechanism is visible in the curve: the Walker2d-6043 student climbs ~600 -> ~6200 as on-policy states are aggregated and the expert labels recovery behaviour. (An under-trained DAgger run with 4 BC epochs/iter looked far worse, the comparison must hold training effort fixed.)
